### 1. What percentage of companies fall under each juridical form?
Why it matters: Understand legal structures preferred in the market.

```python
import sqlite3

dataBase = '/Users/haniehhajighasemi/Documents/BeCode/Bouman9/GitRep/Projects/challenge-data-analysis-sql/kbo_database.db'
connexion = sqlite3.connect(dataBase)
cursor = connexion.cursor()
cursor.execute("""
SELECT 
  e.JuridicalForm,
  c.Description AS JuridicalFormDescription,
  COUNT(*) AS CompanyCount,
  ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (), 2) AS Percentage
FROM enterprise e
LEFT JOIN code c 
  ON CAST(e.JuridicalForm AS TEXT) = c.Code 
  AND c.Category = 'JURIDICAL_FORM' AND c.Language = 'EN'
GROUP BY e.JuridicalForm, c.Description
ORDER BY CompanyCount DESC;

""")
cursor.fetchall()
```

### 4. What are the top sectors (NACE codes) with the most companies?
Why it matters: Shows which industries dominate the market.

```python
import sqlite3

dataBase = '/Users/haniehhajighasemi/Documents/BeCode/Bouman9/GitRep/Projects/challenge-data-analysis-sql/kbo_database.db'
connexion = sqlite3.connect(dataBase)
cursor = connexion.cursor()
cursor.execute("""
SELECT 
  a.NaceCode,
  c.Description AS SectorDescription,
  COUNT(DISTINCT a.EntityNumber) AS CompanyCount
FROM activity a
LEFT JOIN code c 
  ON CAST(a.NaceCode AS TEXT) = c.Code 
  AND c.Category = 'NACE' 
  AND c.Language = 'EN'
GROUP BY a.NaceCode
ORDER BY CompanyCount DESC
LIMIT 10;
""")
cursor.fetchall()
```

### 5. What is the average age of companies by sector?
Why it matters: Shows maturity of different industries.

```python
import sqlite3

dataBase = '/Users/haniehhajighasemi/Documents/BeCode/Bouman9/GitRep/Projects/challenge-data-analysis-sql/kbo_database.db'
connexion = sqlite3.connect(dataBase)
cursor = connexion.cursor()
cursor.execute("""
SELECT 
  a.NaceCode,
  a.Classification AS Sector,
ROUND(
  AVG(
    JULIANDAY('now') - 
    JULIANDAY(
      SUBSTR(e.StartDate, 7, 4) || '-' || SUBSTR(e.StartDate, 4, 2) || '-' || SUBSTR(e.StartDate, 1, 2)
    )
  ) / 365, 
  1
) AS AvgAgeYears
FROM enterprise e
JOIN activity a ON e.EnterpriseNumber = a.EntityNumber
LEFT JOIN code c 
  ON CAST(a.NaceCode AS TEXT) = c.Code 
  AND c.Category = 'NACE' 
  AND c.Language = 'EN'
WHERE e.StartDate IS NOT NULL
GROUP BY a.NaceCode, c.Description
ORDER BY AvgAgeYears DESC;
""")
cursor.fetchall()
```

### 6. Which municipalities have the most active companies?
Why it matters: Highlights economic hotspots.

```python
import sqlite3

dataBase = '/Users/haniehhajighasemi/Documents/BeCode/Bouman9/GitRep/Projects/challenge-data-analysis-sql/kbo_database.db'
connexion = sqlite3.connect(dataBase)
cursor = connexion.cursor()
cursor.execute("""
SELECT 
  MunicipalityNL,
  COUNT(DISTINCT EntityNumber) AS ActiveCompanyCount
FROM address a
JOIN enterprise e ON a.EntityNumber = e.EnterpriseNumber
WHERE e.Status = 'AC'
GROUP BY MunicipalityNL
ORDER BY ActiveCompanyCount DESC
LIMIT 10;
""")
cursor.fetchall()
```

### 7. Which sectors are growing the fastest year-over-year?
Why it matters: Detect booming industries.

```python
import sqlite3

dataBase = '/Users/haniehhajighasemi/Documents/BeCode/Bouman9/GitRep/Projects/challenge-data-analysis-sql/kbo_database.db'
connexion = sqlite3.connect(dataBase)
cursor = connexion.cursor()
cursor.execute("""
SELECT 
  Year,
  NaceCode,
  SectorDescription,
  CompanyCount,
  CompanyCount - LAG(CompanyCount) OVER (PARTITION BY NaceCode ORDER BY Year) AS YoYChange
FROM (
  SELECT 
    STRFTIME('%Y',  SUBSTR(StartDate, 7, 4) || '-' || SUBSTR(StartDate, 4, 2) || '-' || SUBSTR(StartDate, 1, 2)) AS Year,
    a.NaceCode,
    c.Description AS SectorDescription,
    COUNT(*) AS CompanyCount
  FROM enterprise e
  JOIN activity a ON e.EnterpriseNumber = a.EntityNumber
  LEFT JOIN code c 
    ON CAST(a.NaceCode AS TEXT) = c.Code 
    AND c.Category = 'NACE' 
    AND c.Language = 'EN'
  WHERE e.StartDate IS NOT NULL
  GROUP BY Year, a.NaceCode, c.Description
)
ORDER BY YoYChange DESC
LIMIT 15;
""")
cursor.fetchall()
```

### 9. Do companies with more branches tend to be older?
Why it matters: Understand growth and maturity relationship.

```python
import sqlite3

dataBase = '/Users/haniehhajighasemi/Documents/BeCode/Bouman9/GitRep/Projects/challenge-data-analysis-sql/kbo_database.db'
connexion = sqlite3.connect(dataBase)
cursor = connexion.cursor()
cursor.execute("""
SELECT 
  e.EnterpriseNumber,
  COUNT(b.Id) AS BranchCount,
  STRFTIME('%Y', SUBSTR(e.StartDate, 7, 4) || '-' || SUBSTR(e.StartDate, 4, 2) || '-' || SUBSTR(e.StartDate, 1, 2)) AS Year,
  ROUND(
    CASE 
      WHEN e.StartDate IS NOT NULL AND LENGTH(e.StartDate) >= 10 THEN (JULIANDAY('now') - JULIANDAY(SUBSTR(e.StartDate, 7, 4) || '-' || SUBSTR(e.StartDate, 4, 2) || '-' || SUBSTR(e.StartDate, 1, 2))) / 365
      ELSE 0
    END, 
  1) AS CompanyAgeYears
FROM enterprise e
LEFT JOIN branch b ON e.EnterpriseNumber = b.EnterpriseNumber
WHERE e.StartDate IS NOT NULL
GROUP BY e.EnterpriseNumber
HAVING BranchCount > 0
ORDER BY BranchCount DESC
LIMIT 20;

""")
cursor.fetchall()
```
